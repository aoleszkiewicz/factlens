```python
import pandas as pd
from src.paths import DATA_PROCESSED
```

```python
clean = pd.read_csv(DATA_PROCESSED / "news_cleaned.csv")
```

```python
clean.head()
```

```python
from src.paths import DATA_SPLITS
from src.preprocessing.splits import make_splits

train, test, val = make_splits(clean, output_dir=DATA_SPLITS, write_to_file=True, random_state=42)
```

```python
total = len(clean)
print(f"Source rows : {total}")
print(f"train       : {len(train):>6}  ({len(train) / total:.1%})")
print(f"val         : {len(val):>6}  ({len(val) / total:.1%})")
print(f"test        : {len(test):>6}  ({len(test) / total:.1%})")
print(
    f"sum check   : {len(train) + len(val) + len(test)} == {total} → {'OK' if len(train) + len(val) + len(test) == total else 'MISMATCH'}"
)
print()
for name, split in [("train", train), ("val", val), ("test", test)]:
    dist = split["label"].value_counts(normalize=True).sort_index()
    print(f"{name:5}  label=0: {dist[0]:.3f}  label=1: {dist[1]:.3f}")
```

**Wniosek (1).** Funkcja `make_splits` poprawnie dzieli zbiór na części train/val/test w proporcjach 60/20/20. Stratyfikacja zachowuje rozkład klas we wszystkich trzech splitach, a suma wierszy jest równa rozmiarowi zbioru wejściowego.
