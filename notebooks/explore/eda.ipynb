### 1. Ogolny przeglad danych

```python
import kagglehub
import pandas as pd
import plotly.graph_objects as go
```

```python
pd.options.plotting.backend = "plotly"
```

```python
path = kagglehub.dataset_download("clmentbisaillon/fake-and-real-news-dataset")
```

```python
true_df = pd.read_csv(f"{path}/True.csv")
fake_df = pd.read_csv(f"{path}/Fake.csv")

true_df["label"] = 1
fake_df["label"] = 0

news_df = pd.concat([true_df, fake_df], ignore_index=True)
```

```python
print(f"""
true set contains: {true_df.shape[0]} records
fake set contains: {fake_df.shape[0]} records
    
dataset records: ({news_df.shape[0]})
features: ({news_df.shape[1]}) {news_df.columns.tolist()}
""")
```

```python
news_df.info()
```

```python
news_df.describe(include="all")
```

```python
fig = news_df.plot.hist(
    x='label',
    color='label',
    title='Class distribution across the dataset',
    labels={'label': 'Class', 'count': 'Samples'},
    color_discrete_map={0: '#EF553B', 1: "#60D39D"},
    text_auto=True,
    barmode='overlay'
)
fig.update_layout(
    bargap=0.2,
    font=dict(size=13),
    title_font_size=18,
    showlegend=False,
    xaxis=dict(
        tickmode='array',
        tickvals=[0, 1],
        ticktext=['Fake (0)', 'Real (1)']
    )
)

fig.update_traces(textposition='outside')
fig.show()
```

```python
fake_count, true_count = news_df['label'].value_counts()
print(f"Ratio danych (fake/true): {fake_count / true_count}")
```
